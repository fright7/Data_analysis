```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import FinanceDataReader as fdr
```

```python
posco = fdr.DataReader("005490","2021-06")["Close"]
hy_steel = fdr.DataReader("004020","2021-06")["Close"]
dongkuk = fdr.DataReader("001230","2021-06")["Close"]

naver = fdr.DataReader("035420","2021-06")["Close"]
kakao = fdr.DataReader("035720","2021-06")["Close"]
pearl = fdr.DataReader("263750","2021-06")["Close"]
```

```python
sns.set_style("whitegrid")

fig = plt.figure(figsize=(30,10))
ax = fig.subplots(1,2)
ax[0].plot(posco.pct_change(),label="POSCO")
ax[0].plot(hy_steel.pct_change(),label="HY_STEEL")
ax[0].plot(dongkuk.pct_change(),label="DONGKUK STEEL")
ax[0].set_title("Cyclical stock",fontsize=12)

ax[0].axhline(y=0, color='r', linestyle='--')

ax[1].plot(naver.pct_change(),label="NAVER",color="green")
ax[1].plot(kakao.pct_change(),label="KAKAO",color="red")
ax[1].plot(pearl.pct_change(),label="PEARL ABYSS",color="purple")
ax[1].set_title("Growth stock",fontsize=12)

ax[1].axhline(y=0, color='r', linestyle='--')

fig.legend()
```

```python
def cal_avg(stock):

  avg_return = 1

  for i in range(1,len(stock)):

    avg_return *= 1+stock.pct_change()[i]

  return (avg_return**(1/len(stock))-1)*100

#기하평균수익률
```

```python
posco_return = cal_avg(posco)
hy_steel_return = cal_avg(hy_steel)
dongkuk_return = cal_avg(dongkuk)

kakao_return = cal_avg(kakao)
naver_return = cal_avg(naver)
pearl_return = cal_avg(pearl)
```

```python
return_df = pd.DataFrame(
    {
        "POSCO" : [str(round(posco_return,2)) + "%"],
        "HY_STEEL" : [str(round(hy_steel_return,2)) + "%"],
        "DONGKUK STEEL" : [str(round(dongkuk_return,2)) + "%"],
        "KAKAO" : [str(round(kakao_return,2)) + "%"],
        "NAVER" : [str(round(naver_return,2)) + "%"],
        "PEARL_ABYSS" : [str(round(pearl_return,2)) + "%"]

    },
    index=["return"]
)
```
